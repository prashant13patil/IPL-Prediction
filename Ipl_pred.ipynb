```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

input_train_path = "train.csv"
input_test_path = "test.csv"
df_train = pd.read_csv(input_train_path)
df_test = pd.read_csv(input_test_path)
df_train.describe()
```

```python
#df.head(20)
df_train.info()
df_test.info()
```

```python
df_train.shape
```

```python
df_train = df_train[(df_train[['Team 1','Team 2']] != "Airport Flyers").all(axis=1)]
df_train = df_train[(df_train[['Team 1','Team 2']] != "Forum Fans").all(axis=1)]
# df_train = df_train[df_train['Team 1'] != "Airport Flyers" or df_train['Team 2'] != "Airport Flyers"]
df_test = df_test[(df_test[['Team 1','Team 2']] != "HSR High Rent Payers").all(axis=1)]
df_train.shape
```

```python
df_test.loc[8 , 'CityOfGame'] = "Bellandur"
df_test.loc[72 , 'CityOfGame'] = "Koramangala"
```

```python
df_test = df_test.rename(columns={'CityOfGame': 'City', 'Day': 'DayOfWeek'})

winner = []
for index, row in (df_train.iterrows()):
    if (row['Winner (team 1=1, team 2=0)'] == 1):
        winner.append(row['Team 1'])
    else:
        winner.append(row['Team 2'])
df_train['Winner'] = winner
df_train['Winner']

winner = []
for index, row in (df_test.iterrows()):
    if (row['Winner (team 1=1, team 2=0)'] == 1):
        winner.append(row['Team 1'])
    else:
        winner.append(row['Team 2'])
df_test['Winner'] = winner
df_test['Winner']
```

```python
y_train = pd.DataFrame(index = df_train.index)
y_train = df_train[['Winner']]
y_test = pd.DataFrame(index = df_test.index)
y_test = df_test[['Winner']]
train_x = df_train.drop(['Winner (team 1=1, team 2=0)','Game ID','DayOfWeek','DateOfGame','TimeOfGame','Winner'], axis=1)
test_x = df_test.drop(['Winner (team 1=1, team 2=0)','Game ID','DayOfWeek','DateOfGame','TimeOfGame','Winner'], axis=1)

```

```python
print(train_x['Team 2'].unique())
print(test_x['Team 1'].unique())
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

from sklearn.preprocessing import scale

cols =[['AvgWindSpeed', 'AvgHumidity',
       'Inn 1 Team 1 NOP R>25,SR>125', 'Inn 1 Team 1 NOP R<25, SR>125',
       'Inn 1 Team 1 Total 4s', 'Inn 1 Team 1 Total 6s',
       'Inn 1 Team 1 Max Strike Rate_ALLBatsmen',
       'Inn 1 Team 2 NoP fast bowlers', 'Inn 1 Team 2 NoP Spinners',
       'Inn 1 Team 2 wickets taken_catches_runout',
       'Inn1 Team 2 wickets taken_ bowled _lbw_caught by keeper_stumping',
       'Inn 1 Team 2 Extras conceded in_wides_No Balls',
       'Inn 2 Team 2 NOP R>25,SR>125', 'Inn 2 Team 2 NOP R<25, SR>125',
       'Inn 2 Team 2 Total 4s', 'Inn 2 Team 2 Total 6s',
       'Inn 2 Team 2 Max Strike Rate_ALLBatsmen',
       'Inn 2 Team 1 NoP fast bowlers', 'Inn 2 Team 1 NoP Spinners',
       'Inn 2 Team 1 wickets taken_catches_runout',
       'Inn2 Team 1 wickets taken_ bowled _lbw_caught by keeper_stumping',
       'Inn 2 Team 1 Extras conceded in_wides_No Balls']]

for col in cols:
    train_x[col] = scale(train_x[col])
    test_x[col] = scale(test_x[col])
```

```python
import re
regex = re.compile(r"\[|\]|<", re.IGNORECASE)

final_train_x = pd.DataFrame(index = train_x.index)
final_test_x = pd.DataFrame(index = test_x.index)


for col, col_data in train_x.iteritems():
    if col_data.dtype == object:
        col_data = pd.get_dummies(col_data, prefix = col)
    final_train_x = final_train_x.join(col_data)
    
for col, col_data in test_x.iteritems():
    if col_data.dtype == object:
        col_data = pd.get_dummies(col_data, prefix = col)
    final_test_x = final_test_x.join(col_data)

final_train_x.columns = [regex.sub("_", col) if any(x in str(col) for x in set(('[', ']', '<'))) else col for col in final_train_x.columns.values]
final_test_x.columns = [regex.sub("_", col) if any(x in str(col) for x in set(('[', ']', '<'))) else col for col in final_test_x.columns.values]
print(final_test_x)

```
